### Generación de dataset para R-Studio

Para generar el dataset que se usara para generar el mapa en el Shiny dashboard, es necesario hacer una agrupacion de datos y una modificacion para hacerlo y observar el numero total de casos en la __Republica Mexica hasta el 31 de Enero del 2021__.

### Modificación del DataSet para ver si murio o no murio la persona

Con la columna `FECHA_DEF` nos damos idea si la persona fallecio o no. 


Si la fecha es __9999-99-99__ significa que fallecio y por lo tanto se sustituira con un __1__, que significa que _ Murio_. En otro caso se sustituira con un __2__.

```python
#Se hace copia del dataset
df2 = df1.copy()
```

```python
# Exploracion de los datos de "FECHA_DEF"
df2.iloc[:,12].value_counts()
```

```python
# Se sustituyen los valores que no son 9999-99-99 por 1
df2['FECHA_DEF'] = df2['FECHA_DEF'].map(lambda x : x if x == '9999-99-99' else 1).copy()

# Se sustituyen los valores que son 9999-99-99 por 2
df2['FECHA_DEF'] = df2['FECHA_DEF'].replace({'9999-99-99':2})

# Se renombra la columna de fecha de desenso por si esta vivo o no 
df2.rename(columns={'FECHA_DEF':'ESTADO_PACIENTE'})

# Observacion de los cambios
df2['FECHA_DEF'].value_counts()
```
